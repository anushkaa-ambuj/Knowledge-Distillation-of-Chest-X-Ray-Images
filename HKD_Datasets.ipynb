#**Visualizing Datasets**

```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

## Fashion MNIST

```python
# Load the Fashion MNIST dataset
train_dataset = torchvision.datasets.FashionMNIST(root='./data', train=True, transform=torchvision.transforms.ToTensor(), download=True)
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
import random
# Visualize 10 random images from each class
num_samples_per_class = 10
fig, axes = plt.subplots(10, num_samples_per_class, figsize=(12, 12))

for i, ax_row in enumerate(axes):
    class_indices = torch.where(train_dataset.targets == i)[0]
    random_indices = random.sample(class_indices.tolist(), num_samples_per_class)

    for j, idx in enumerate(random_indices):
        image, _ = train_dataset[idx]
        ax = ax_row[j]
        ax.imshow(image.squeeze(), cmap='gray')
        ax.axis('off')
        ax.set_title(class_names[i])

plt.tight_layout()
plt.show()
```

## CIFAR10

```python
# Load the CIFAR-10 dataset and define class names
transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor(),])

train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, transform=transform, download=True)
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# Visualize 10 random images from each class
num_samples_per_class = 10
fig, axes = plt.subplots(10, num_samples_per_class, figsize=(12, 12))

for i, ax_row in enumerate(axes):
    class_indices = [idx for idx, target in enumerate(train_dataset.targets) if target == i]
    random_indices = random.sample(class_indices, num_samples_per_class)

    for j, idx in enumerate(random_indices):
        image, _ = train_dataset[idx]
        ax = ax_row[j]
        image = np.transpose(image.numpy(), (1, 2, 0))  # Convert to (H, W, C) format
        ax.imshow(image)
        ax.axis('off')
        ax.set_title(class_names[i])

plt.tight_layout()
plt.show()
```
